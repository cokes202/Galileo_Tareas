## Creando la Funcion 
### y = mx + b

```python
def regresion(varindp,y,epochs,imprimir,lr):
    munos = np.ones_like(varindp)
    varyunos = np.hstack((varindp,munos))
    m = 0
    b = 0
    cont = 1
    coeficientes =  np.array([[m],
                             [b]])
    learningr = lr 
    lerror = [] 
    dicparametros = {}
    for cont in range(epochs):
        y_hat = np.matmul(varyunos,coeficientes)
        error = (np.mean(np.power(y-y_hat,2)))*1/2
        resta = y_hat - y
        restaa = np.squeeze(np.asarray(np.hstack((resta,resta))))
        gradientes = np.reshape(np.mean(restaa * varyunos,axis=0),(2,1))        
        coeficientes = coeficientes - (learningr*gradientes)             
        lerror.append(error)        
        dicparametros[cont] = coeficientes
        dicparametros[0] = varyunos
        
        if cont % imprimir == 0: 
            print("Error", cont,error) 
                
    return dicparametros,lerror
```

### Usando la funcion con la Variable Quality

```python
estimadorql,errorql = regresion(quality,saleprice,100,10,.001)

print(estimadorql)
print(errorql)

print(type(estimadorql))
print(type(errorql))
```

## Función para Graficar los errores de la Funcion de Regresión

```python
def graficaerrores(x):
    arregloer = np.reshape(np.array(x),(len(x),1))
    iteraciones = np.reshape(np.arange(1,len(arregloer)+1),(len(x),1))
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(iteraciones, arregloer, color="black")
    ax.set_xlabel("Paso Iteración")
    ax.set_ylabel("Error Observado")
    ax.set_title("Error variable independiente")
    plt.show()
```

## Función Unificada

```python
def modeloestimado(z,mb,x):
    diccionario = z
    fin = len(diccionario)
    munos = np.ones_like(x)
    varyunos = np.hstack((x,munos))
    
    mb_manual =  diccionario[fin-1]
    pronmanual = np.matmul(varyunos,mb_manual)
    pronscikit = np.matmul(varyunos,mb)
    pronpromedio =  (pronmanual +  pronscikit)*1/2
    
    return pronmanual, pronscikit, pronpromedio
```
